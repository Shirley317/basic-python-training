### 练习1.4
对元组做相同练习，并比较列表和元组的读取时间差异


````python
#练习1.3 打印二维列表
data_list = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]

# 方法1：一次打印完整内容
def print_list1(data):
    for row in data:
        print(*row, end=' ')
    print()    
%time print_list1(data_list)

# 方法2：外循环遍历行，内循环遍历列
def print_list2(data):
    for row in data:
        for item in row:
            print(item, end=' ')
        print()        
%time print_list2(data_list)
print(" ")
#练习1.4 打印二维元组
data_tuple = ((1, 2, 3), (4, 5, 6), (7, 8, 9))

# 方法1：一次打印完整内容
def print_tuple1(data):
    for row in data:
        print(*row, end=' ')
    print()    
%time print_tuple1(data_tuple) 

# 方法2：外循环遍历行，内循环遍历列
def print_tuple2(data):
    for row in data:
        for item in row:
            print(item, end=' ')
        print()
%time print_tuple2(data_tuple)

#比较分析：元组的读取时间少于列表
````

②将其中一组余弦函数绘制为shade背景。

````python
import numpy as np
import matplotlib.pyplot as plt

# 定义数据
x = np.linspace(0, 2 * np.pi, 1000) 
y1 = np.cos(x)
y2 = 0.5 * np.cos(x + np.pi / 4)
y3 = 1.5 * np.cos(x - np.pi / 6)

# 创建图形
plt.figure(figsize=(10, 6))

# 绘制三条不同的余弦函数，测试透视度、线条粗细、线条种类、颜色
plt.plot(x, y1, label=r'$y = \cos(x)$', color='blue', linestyle='-', linewidth=2, alpha=0.7)
plt.plot(x, y2, label=r'$y = 0.5 \cos(x + \frac{\pi}{4})$', color='green', linestyle='--', linewidth=2, alpha=0.7)
plt.plot(x, y3, label=r'$y = 1.5 \cos(x - \frac{\pi}{6})$', color='red', linestyle='-.', linewidth=2, alpha=0.7)

# 使用 LaTeX 数学公式标注
plt.title(r'Cosine of different phases and amplitudes', fontsize=16)
plt.xlabel(r'$x$', fontsize=14)
plt.ylabel(r'$y$', fontsize=14)

# 添加注释
plt.annotate(r'$y = \cos(x)$', xy=(np.pi, 0), xytext=(np.pi, 0.5),
             arrowprops=dict(facecolor='black', shrink=0.05))

# 添加图例
plt.legend()

# 将其中一组余弦函数绘制为阴影背景
plt.fill_between(x, y2, color='green', alpha=0.1)

# 显示图形
plt.grid(True)
plt.show()
````

````python
import numpy as np 
import matplotlib.pyplot as plt

def my_func(x):
    return x**2-2*x

def grad_func(x):
    return 2*x-2

eta=0.1 #学习参数
x=4     #初始值 应该给范围内的随机数
record_x=[]  #x的记录
record_y=[]    #y的记录
for i in range(20): #将x更新20次
    y=my_func(x)
    record_x.append(x)
    record_y.append(y)
    x-=eta*grad_func(x) # x=x-eta*grad_func(x)
    
x=np.linspace(-2,4)
y=my_func(x)

plt.plot(x,y,linestyle="dashed") #函数图
plt.scatter(record_x,record_y) #散点图

plt.xlabel("x",size=14)
plt.ylabel("y",size=14)
plt.grid()
plt.show()
````

### 练习4.2 用梯度下降法来计算函数$f(x)=x^4+2x^3-3x^2-2x$的局部和整体最小值

注意：设计算法计算，不可以人工分段
提示：可用修正梯度下降法

````python
# 修正梯度下降法（Modified Gradient Descent）通过引入动量来加速收敛
import numpy as np
import matplotlib.pyplot as plt

# 定义目标函数
def f(x):
    return x**4 + 2*x**3 - 3*x**2 - 2*x

# 定义目标函数的导数
def df(x):
    return 4*x**3 + 6*x**2 - 6*x - 2

# 修正梯度下降法（带动量）
def modified_gradient_descent(df, x_init, alpha, beta, max_iter):
    x = x_init
    v = 0  # 初始化动量
    for i in range(max_iter):
        grad = df(x)
        v = beta * v + alpha * grad  # 计算新的动量，beta 是动量系数。
        x_new = x - v  # 更新 x
        if np.abs(x_new - x) < 1e-6:  # 收敛条件 np.abs(x_new - x) 表示两者的绝对差
            break
        x = x_new
    return x

# 初始化参数
alpha = 0.01      #学习率
beta = 0.9        #动量参数
max_iter = 10000  #最大迭代次数

# 找不同初始值下的局部最小值
initial_values = np.linspace(-3, 3, 10)  # 生成从 -3 到 3 的 10 个初始值
minima = []

for x_init in initial_values:
    min_x = modified_gradient_descent(df, x_init, alpha, beta, max_iter)
    minima.append((x_init, min_x, f(min_x)))

# 找到最小的局部最小值，认为是全局最小值
minima.sort(key=lambda x: x[2])  # 按 f(min_x) 排序
global_min_x = minima[0][1]
global_min_f = minima[0][2]

# 打印局部最小值和全局最小值
for init, min_x, min_f in minima:
    print(f"Initial x: {init:.2f}, Local minimum x: {min_x:.6f}, f(x): {min_f:.6f}")
print(f"\nGlobal minimum x: {global_min_x:.6f}, f(x): {global_min_f:.6f}")

# 绘制目标函数
x_vals = np.linspace(-3, 3, 400)
y_vals = f(x_vals)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label=r'$f\,(x) = x^4 + 2x^3 - 3x^2 - 2x$') #\,表示空格
plt.xlabel("x", size=14)
plt.ylabel("f (x)", size=14)
plt.title(r'Plot of the function $f\,(x) = x^4 + 2x^3 - 3x^2 - 2x$', size=16)
plt.legend()
plt.grid(True)

# 绘制局部最小值和全局最小值
for init, min_x, min_f in minima:
    plt.plot(min_x, min_f, 'ro')  # 标记局部最小值,'ro'红点
    plt.text(min_x, min_f, f"({min_x:.2f}, {min_f:.2f})", fontsize=12)
    
plt.plot(global_min_x, global_min_f, 'c^', markersize=15)  # 标记全局最小值
plt.text(global_min_x, global_min_f + 8, "Global min", fontsize=14, color='Turquoise', ha='center')
plt.show()
# plt.text()用于在图形上添加文本注释, ha 是horizontal alignment的缩写，表示水平对齐方式
````

### 练习5.3 计算方阵的逆矩阵linalg.inv()和转置.T，并验证是否单位矩阵相等eye()

````python
import numpy as np

a= np.array([[1, 2], [3, 4]])
inverse_a = np.linalg.inv(a) #计算的必须是方阵
transpose_a = a.T
result_matrix = np.dot(inverse_a, a)
#np.eye()此函数生成一个二维数组，其中对角线上为 1，其他位置为 0; a.shape[0]：这将获取原始矩阵的行数。
identity_matrix = np.eye(a.shape[0])
is_identity = np.allclose(result_matrix, identity_matrix)

print("原矩阵：\n", a)
print("逆矩阵：\n", inverse_a)
print("转置矩阵：\n", transpose_a)
print("逆矩阵乘原矩阵：\n", result_matrix)
print("是否等于单位矩阵：", is_identity)
````

### 练习5.4 绘制向量与标准基底的图像

① 计算并绘制向量的基向量；
② 利用旋转矩阵将向量旋转逆时针45度角，并展示结果。


````python
import numpy as np
import matplotlib.pyplot as plt

# Original vector
a = np.array([2, 3])

# Basis vectors
a_x = np.array([2, 0])
a_y = np.array([0, 3])

# Rotation matrix for 45 degrees counterclockwise
theta = np.pi / 4  # 45 degrees in radians
rotation_matrix = np.array([[np.cos(theta), -np.sin(theta)],
                            [np.sin(theta), np.cos(theta)]])

# Rotated vector
rotated_a = np.dot(rotation_matrix, a)

def arrow(start,size,color):
    plt.quiver(start[0],start[1],size[0],size[1],angles="xy",scale_units="xy",scale=1,color=color)
    
s=np.array([0,0])
arrow(s,a,color="black")
arrow(s,a_x,color="blue")
arrow(s,a_y,color="blue")
arrow(s,rotated_a,color="red")

plt.xlim([-3,3])
plt.ylim([-1,4])
plt.xlabel("x",size=14)
plt.ylabel("y",size=14)
plt.grid()
plt.show()
````

程序完成之后，优化执行效率是大型计算程序的重要环节。 

- ``%time``: 监视执行一行命令的时长，执行一次。
- ``%timeit``: 监视执行一行命令的时长，重复多次以获得更加准确的结果。 
- ``%prun``: 利用性能分析器运行程序。
- ``%lprun``: 利用性能分析器逐行运行程序。 
- ``%memit``: 测试一个命令所占内存。 
- ``%mprun``: 利用内存性能分析器逐行运行程序。 

``line_profiler`` 和 ``memory_profiler`` 包含以上部分命令。

### 7.2 性能分析``%prun``

````python
def sum_of_lists(N):
    total = 0
    for i in range(5):
        L = [j ^ (j >> i) for j in range(N)]
        total += sum(L)
    return total
````

````python
%prun sum_of_lists(1000000)
````

结果显示了每个函数所耗费的资源。
